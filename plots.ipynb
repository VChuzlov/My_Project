```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
u = np.loadtxt(r'..\Distillation2\xi_backup.txt')
```

```python
yu = np.loadtxt(r'..\Distillation2\yi_backup.txt')
yc = np.loadtxt(r'..\Distillation2\yi.txt')
```

```python
fig, ax = plt.subplots()

ax.plot(u[:, 2], marker='o', label='C3-u')
ax.plot(c[:, 2], marker='d', label='C3-c')

ax.plot(u[:, 4], marker='o', label='nC4-u')
ax.plot(c[:, 4], marker='d', label='nC4-c')

ax.plot(u[:, 5], marker='o', label='iC5-u')
ax.plot(c[:, 5], marker='d', label='iC5-c')

ax.plot(u[:, 6], marker='o', label='nC5-u')
ax.plot(c[:, 6], marker='d', label='nC5-c')

plt.legend(ncol=4, title='Liquid mole fractions')
plt.tight_layout()
plt.savefig('xij.pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(yu[:, 2], marker='o', label='C3-u')
ax.plot(yc[:, 2], marker='d', label='C3-c')

ax.plot(yu[:, 4], marker='o', label='nC4-u')
ax.plot(yc[:, 4], marker='d', label='nC4-c')

ax.plot(yu[:, 5], marker='o', label='iC5-u')
ax.plot(yc[:, 5], marker='d', label='iC5-c')

ax.plot(yu[:, 6], marker='o', label='nC5-u')
ax.plot(yc[:, 6], marker='d', label='nC5-c')

plt.legend(ncol=4, title='Vapor mole fractions')
plt.tight_layout()
plt.savefig('yij.pdf')
```

```python
flows = np.loadtxt(r'..\Distillation2\profile_backup.txt')
t = np.loadtxt(r'..\Distillation2\t_backup.txt')
```

```python
profiles_c = np.loadtxt(r'..\Distillation2\profiles.txt')
t_c = np.loadtxt(r'..\Distillation2\t.txt')
```

```python
fig, ax = plt.subplots()

ax.plot(flows[:, 0], marker='o', label='Liquid-U')
ax.plot(profiles_c[:, 0], marker='d', label='Liquid-C')

plt.legend()
plt.tight_layout()
plt.savefig('liquid.pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(flows[:, 1], marker='o', label='Vapor-U')
ax.plot(profiles_c[:, 1], marker='d', label='Vapor-C')

plt.legend()
plt.tight_layout()
plt.savefig('vapor.pdf')
```

```python
fig, ax = plt.subplots()

ax.plot(t, marker='o', label='Temperature-U')
ax.plot(t_c, marker='d', label='Temperature-C')

plt.legend()
plt.tight_layout()
plt.savefig('temperature.pdf')
```
