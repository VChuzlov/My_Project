```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.loadtxt(r'ul.txt')
```

```python
c = np.loadtxt(r'..\Distillation2\xi.txt')
```

```python
c.shape, u.shape
```

```python
fig, ax = plt.subplots()

ax.plot(u[:, 2], marker='o', label='C3-u')
ax.plot(c[:, 2], marker='d', label='C3-c')

ax.plot(u[:, 4], marker='o', label='C4-u')
ax.plot(c[:, 4], marker='d', label='C4-c')

ax.plot(u[:, 5], marker='o', label='C5-u')
ax.plot(c[:, 5], marker='d', label='C5-c')

plt.legend(ncol=3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()

for i, _ in enumerate(u):
    ax.scatter(u[i], c[i])

ax.plot([0, 1], [0, 1], 'k--', alpha=.5)

plt.tight_layout()
```
