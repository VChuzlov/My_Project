```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.loadtxt(r'ul.txt')
```

```python
c = np.loadtxt(r'..\Distillation2\xi.txt')
```

```python
c.shape, u.shape
```

```python
fig, ax = plt.subplots()

ax.plot(u[:, -1], marker='o', label='C8-u')
ax.plot(c[:, -1], marker='d', label='C8-c')

ax.plot(u[:, -2], marker='o', label='C7-u')
ax.plot(c[:, -2], marker='d', label='C7-c')

ax.plot(u[:, -3], marker='o', label='C6-u')
ax.plot(c[:, -3], marker='d', label='C6-c')

plt.legend(ncol=3)
plt.tight_layout()
```

```python
fig, ax = plt.subplots()

for i, _ in enumerate(u):
    ax.scatter(u[i], c[i])

ax.plot([0, .6], [0, .6], 'k--', alpha=.5)

plt.tight_layout()
```
